```python
# Comparing the race of exonerated prisoners
rd = df['race'].value_counts()
colors = ['#d9534f', '#5bc0de', '#88d8b0', '#428bca', '#5cb85c', '#ffcc5c', '#f47835']
trace = go.Pie(labels=list(rd.index), values=list(rd.values), textinfo='label+percent', 
               textfont=dict(size=16),
               marker=dict(colors=colors, 
                           line=dict(color='#000000', width=1)))
py.iplot([trace], filename='exonerations_by_race', colors=colors)
```

```python
# Compared to the prison populations race (from Federal Bureau of Prisons on September 23, 2017)
colors = ['#d9534f', '#5bc0de', '#88d8b0', '#5cb85c', '#ffcc5c']
races = ['Black', 'Caucasian', 'Hispanic', 'Asian', 'Native American']
prison_pop = [70294, 59734, 48676, 4096, 2707]
trace = go.Pie(labels=races, 
               values=prison_pop,
               textinfo='label+percent', 
               textfont=dict(size=16),
               marker=dict(colors=colors, 
                           line=dict(color='#000000', width=1)))
py.iplot([trace], filename='prisoners_by_race')
```

```python
# Compared to US population by percent (from US census)
us_pop = [13.3, 76.9, 5.7, 1.3, 17.8]
trace = go.Pie(labels=races, 
               values=us_pop, 
               textinfo='label+percent', 
               textfont=dict(size=16),
               marker=dict(colors=colors, 
                           line=dict(color='#000000', width=1)))
py.iplot([trace], filename='us_population')
```

```python
rd_new = rd.drop(['Don\'t Know', 'Other'])

fig = {
    'data': [
        {
            'labels': list(rd_new.index),
            'values': list(rd_new.values),
            'type': 'pie',
            'name': 'Exonerations',
            'marker': {'colors': colors},
            'domain': {'x': [0, .48],
                       'y': [0, .49]},
            'hoverinfo':'name',
            'textinfo':'label+percent'
        },
        {
            'labels': races,
            'values': prison_pop,
            'marker': {'colors': colors},
            'type': 'pie',
            'name': 'Prison Population',
            'domain': {'x': [.52, 1],
                       'y': [0, .49]},
            'hoverinfo':'name',
            'textinfo':'label+percent'

        },
        {
            'labels': races,
            'values': us_pop,
            'marker': {'colors': colors},
            'type': 'pie',
            'name': 'US pop.',
            'domain': {'x': [0, .48],
                       'y': [.51, 1]},
            'hoverinfo':'name',
            'textinfo':'label+percent'
        }
    ],
    'layout': {'title': 'Comparing Exoneration Rates, Prison Populations, and US Population by Race',
               'showlegend': False}
}

py.iplot(fig, filename='race_comparison_subplots')
```

```python
edf = pd.Series([df['mistake_witness'].count(), 
                          df['false_confession'].count(), 
                          df['perjury'].count(),
                          df['false_evidence'].count(),
                          df['official_misconduct'].count(),
                          df['inadequate_defense'].count()])
causes = ['mistake_witness', 'false_confession', 'perjury', 'false_evidence', 'official_misconduct', 'inadequate_defense']
data = [go.Bar(
            x=causes,
            y=list(edf.values)
    )]
layout = go.Layout(
    title='Reason Conviction was Reversed',
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='exoneration-cause')
```
